```python
# imports
import pandas as pd
```

```python
HGMD_lofgof_df = pd.read_csv('/Users/sayehkohani/Data/goflof_HGMD2019_v032021.csv')
ClinVar_lofgof_df = pd.read_csv('/Users/sayehkohani/Data/goflof_ClinVar_v062021.csv')
```

```python
#Isolate all GOF 
HGMD_gof_df = HGMD_lofgof_df.loc[HGMD_lofgof_df['LABEL'] == 'GOF']
ClinVar_gof_df = ClinVar_lofgof_df.loc[ClinVar_lofgof_df['LABEL'] == 'GOF']
```

```python
HGMD_gof_df.to_csv('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/All_GOF_HGMD.txt', sep='\t')
ClinVar_gof_df.to_csv('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/All_GOF_ClinVar.txt', sep='\t')
```

```python
# Calculate number of appearing GOF variants per gene
gof_per_gene_HGMD_dict = HGMD_gof_df['GENE'].value_counts().to_dict()
gof_per_gene_ClinVar_dict = ClinVar_gof_df['GeneSymbol'].value_counts().to_dict()
```

```python
gof_per_gene_HGMD_df = pd.DataFrame(list(gof_per_gene_HGMD_dict.items()), columns = ['GENE','#GOF'])
gof_per_gene_ClinVar_df = pd.DataFrame(list(gof_per_gene_ClinVar_dict.items()), columns = ['GENE','#GOF'])
```

```python
# save

gof_per_gene_ClinVar_df.to_excel('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/GOF_Per_Gene_ClinVar.xlsx')
gof_per_gene_HGMD_df.to_excel('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/GOF_Per_Gene_HGMD.xlsx')
```

```python
# Merge ClinVar and HGMD data

gof_per_gene_ClinVar_HGMD_df = pd.merge(gof_per_gene_HGMD_df, gof_per_gene_ClinVar_df, how = 'outer', on = 'GENE', suffixes=('_HGMD', '_ClinVar'))
gof_per_gene_ClinVar_HGMD_df
```

```python
gof_per_gene_ClinVar_HGMD_df.to_excel('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/GOF_Per_Gene_HGMD_ClinVar.xlsx')
```
