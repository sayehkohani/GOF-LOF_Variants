```python
import pandas as pd
import gzip
```

```python
column_names = ['Var_ID_hg38', 'Gencode_ID', 'Gene_Symbol', 'Var_Type']
bamovar_df = pd.read_csv('/Users/sayehkohani/Data/GOF/BamoVar/extracted_columns_varid.txt', sep = '\t', names=column_names)
```

```python
bamovar_df.head()
```

```python
logofunc_predictions_file = gzip.open('/Users/sayehkohani/Data/GOF/logofunc-predictions-main/LoGoFunc_preds_SK_updated.csv.gz', 'rt')
```

```python
logofunc_predictions_df = pd.read_csv(logofunc_predictions_file, sep='\t', header = 0)
```

```python
logofunc_predictions_df.head()
```

```python
logofunc_predictions_df = logofunc_predictions_df.drop('Coord', axis = 1)
```

```python
column_name_mapping = {
    'Pos_hg19': 'Pos_hg38',
    'Var_ID_hg19': 'Var_ID_hg38',
    # Add more columns as needed
}

logofunc_predictions_df = logofunc_predictions_df.rename(columns = column_name_mapping)
```

```python
logofunc_predictions_df.head()
```

```python
logofunc_predictions_bamovar = pd.merge(logofunc_predictions_df, bamovar_df, how = 'left', on = 'Var_ID_hg38')
```

```python
logofunc_predictions_bamovar.head()
```

```python
desired_column_order = ['Chr', 'Pos_hg38', 'REF', 'ALT', 'Var_ID_hg38', 'Gencode_ID', 'Gene_Symbol','Var_Type', 'prediction','LoGoFunc_neutral','LoGoFunc_GOF','LoGoFunc_LOF']
logofunc_predictions_bamovar = logofunc_predictions_bamovar[desired_column_order]
```

```python
logofunc_predictions_bamovar.head()
```

```python
# Save logofunc_bamovar_annotated file
logofunc_predictions_bamovar.to_csv('/Users/sayehkohani/Data/GOF/logofunc-predictions-main/LoGoFunc_preds_SK_bamovar.csv.gz', sep = '\t', index=False, compression='gzip')
```

## Isolate all GOF

```python
all_gof_df = logofunc_predictions_bamovar[logofunc_predictions_bamovar['prediction'] == 'GOF']
```

```python
# Find top genes
gof_per_gene_dict = all_gof_df['Gene_Symbol'].value_counts().to_dict()
gof_per_gene_df = pd.DataFrame(list(gof_per_gene_dict.items()), columns = ['Gene','#GOF_LoGoFunc'])
```

```python
gof_per_gene_df
```

```python
gof_per_gene_df.to_csv('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/All_GOF_LoGoFunc.txt', sep = '\t')
```

```python
# Merge with the annotated file

annotated_all_gof = pd.read_excel('/Users/sayehkohani/Dropbox/Research/Sanders_Lab/GoF/Output/Annotated_GOF_Per_Gene_HGMD_ClinVar.xlsx')
```

```python
annotated_all_gof.head()
```

```python
merged_all_gof = pd.merge(gof_per_gene_df, annotated_all_gof, on = 'Gene', how = 'outer')
```
