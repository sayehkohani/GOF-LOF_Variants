## Identify breakdown of LOF/GOF/Neutral

```python
# Calculate value counts
value_counts_logofunc = logofunc_predictions_df['prediction'].value_counts()
value_counts_HGMD = HGMD_df['LABEL'].value_counts()
value_counts_ClinVar = ClinVar_df['LABEL'].value_counts()

# Calculate percentages
total_count_logofunc = value_counts_logofunc.sum()
total_count_HGMD = value_counts_HGMD.sum()
total_count_ClinVar = value_counts_ClinVar.sum()

percentages_logofunc = (value_counts_logofunc / total_count_logofunc) * 100
percentages_HGMD = (value_counts_HGMD / total_count_HGMD) * 100
percentages_ClinVar = (value_counts_ClinVar / total_count_ClinVar) * 100
```

```python
# Separate the main dataframe into sub-dataframes
main_columns = ['GOF', 'LOF', 'Neutral']
disease_columns = panel_df.columns.difference(main_columns)

main_df = panel_df[main_columns]
disease_df = panel_df[disease_columns]
```

```python
# Create a new dataframe for the results
result_df = pd.DataFrame()

# Perform element-wise multiplication using broadcasting
for col in main_df.columns:
    result = main_df[col].values[:, np.newaxis] * disease_df.values
    result_df = pd.concat([result_df, pd.DataFrame(result)], axis=1)

# Rename the columns of the result dataframe
result_df.columns = [f"{col}_{disease}" for col in main_df.columns for disease in disease_df.columns]
```

```python
result_df.drop(columns=['GOF_Panel_Haematological disorders','Neutral_Panel_Haematological disorders','LOF_Panel_Haematological disorders', 'GOF_Panel_Other','LOF_Panel_Other','Neutral_Panel_Other','GOF_Panel_Viral research','LOF_Panel_Viral research','Neutral_Panel_Viral research'], inplace=True)
```

```python
result_df
```

```python
gof_df = result_df.iloc[:, :18]
lof_df = result_df.iloc[:, 18:2 * 18]
neutral_df = result_df.iloc[:, 2 * 18:]
```

```python
gof_df
```

```python
gof_counts = gof_df.sum()
lof_counts = lof_df.sum()
neutral_counts = neutral_df.sum()
```

```python
# Calculate the percentage
gof_percentage_dict = {title: (total / gof_counts.sum()) * 100 for title, total in gof_counts.items()}
lof_percentage_dict = {title: (total / lof_counts.sum()) * 100 for title, total in lof_counts.items()}
neutral_percentage_dict = {title: (total / neutral_counts.sum()) * 100 for title, total in neutral_counts.items()}

# Remove the consequence from key
gof_percentage_d = {}
lof_percentage_d = {}
neutral_percentage_d = {}

substring_to_remove = 'GOF_'
for key, value in gof_percentage_dict.items():
    new_key = key.replace(substring_to_remove, '')
    gof_percentage_d[new_key] = value


substring_to_remove = 'LOF_'
for key, value in lof_percentage_dict.items():
    new_key = key.replace(substring_to_remove, '')
    lof_percentage_d[new_key] = value

substring_to_remove = 'Neutral_'
for key, value in neutral_percentage_dict.items():
    new_key = key.replace(substring_to_remove, '')
    neutral_percentage_d[new_key] = value
    
```

```python
combined_dict = {}

for key in gof_percentage_d:
    combined_dict[key] = [lof_percentage_d[key], gof_percentage_d[key], neutral_percentage_d[key]]

combined_dict
```

```python
sorted_combined_dict = dict(sorted(combined_dict.items(), key=lambda item: max(item[1]), reverse=True))
```
