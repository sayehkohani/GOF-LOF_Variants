```python
all_variants_SCN2A_hg38
```

```python
all_variants_SCN2A_hg38['Var_ID_hg38'] = ('chr'+
    all_variants_SCN2A_hg38['Chr'].astype(str) + ':' +
    all_variants_SCN2A_hg38['Pos_hg38'].astype(str) + ':' +
    all_variants_SCN2A_hg38['Ref'].astype(str) + ':' +
    all_variants_SCN2A_hg38['Alt'].astype(str))
```

```python
all_variants_SCN2A_hg38
```

```python
logofunc_SCN2A = logofunc_predictions_df[logofunc_predictions_df['Gene_Symbol'] == 'SCN2A']
```

```python
logofunc_SCN2A
```

```python
SCN2A_merged = pd.merge(logofunc_SCN2A, all_variants_SCN2A_hg38, on = 'Var_ID_hg38', how = 'inner')
```

```python
SCN2A_merged['Classification'].values.tolist()
```

```python
def determine_pred_cat(classification):
    if classification in ['IEE', 'BIS', 'IEE_Mild/Ataxia']:
        return 'GOF'
    elif classification in ['ASD/DD', 'Schizophrenia']:
        return 'LOF'
    else:
        return ''

# Apply the function to create the 'Pred_cat' column
SCN2A_merged['Pred_cat'] = SCN2A_merged['Classification'].apply(determine_pred_cat)
```

```python
SCN2A_merged
```

```python
SCN2A_merged = SCN2A_merged[SCN2A_merged['Pred_cat'] != '']
```

```python
total = SCN2A_merged.shape[0]
correct = ((SCN2A_merged["prediction"] == "LOF") & (SCN2A_merged["Pred_cat"] == "LOF") |
         (SCN2A_merged["prediction"] == "GOF") & (SCN2A_merged["Pred_cat"] == "GOF")).sum()

error = (total-correct)/total*100
print(error)
```

```python
error
```

```python
act_LOF = len(SCN2A_merged[SCN2A_merged['Pred_cat'] == 'LOF'])
pre_LOF = len(SCN2A_merged[SCN2A_merged['prediction'] == 'LOF'])

act_GOF = len(SCN2A_merged[SCN2A_merged['Pred_cat'] == 'GOF'])
pre_GOF = len(SCN2A_merged[SCN2A_merged['prediction'] == 'GOF'])

categories = ['LOF', 'GOF']
actual = [act_LOF, act_GOF]
predicted = [pre_LOF, pre_GOF]
```

```python
actual
```

```python
predicted
```

```python
bar_width = 0.2  # Width of each bar
bar_positions_set1 = range(len(categories))
bar_positions_set2 = [pos + bar_width for pos in bar_positions_set1]


plt.style.use("seaborn-muted")

# Create a figure and axis with white background
fig, ax = plt.subplots()

# Create horizontal bar charts
ax.barh(bar_positions_set1, actual, height=bar_width, label='Experimental')
ax.barh(bar_positions_set2, predicted, height=bar_width, label='Predicted')

ax.set_xlabel('Number of Variants')
ax.legend()

# Set y-axis ticks and labels to categories
ax.set_yticks([pos + bar_width for pos in bar_positions_set1])
ax.set_yticklabels(categories)

# Remove y-axis tick marks
ax.tick_params(axis='y', which='both', length=0)

plt.savefig('/Users/sayehkohani/Downloads/SCN2A_Logofunc.png', bbox_inches='tight')  # Change the filename and format as needed

plt.show()
```

```python
total = PTEN_merged_up.shape[0]
correct = (
    ((PTEN_merged_up["prediction"] == "Neutral") & (PTEN_merged_up['Polyphen2_HVAR_pred'] == 'B')) |
    ((PTEN_merged_up["prediction"] == "LOF") &
    (PTEN_merged_up["Polyphen2_HVAR_pred"].isin(["D", "P"])))
).sum()

error = (total-correct)/total*100
print(error)
```

```python
total
```

```python
act_neutral = len(PTEN_merged_up[PTEN_merged_up['Polyphen2_HVAR_pred'] == 'B'])
pre_neutral = len(PTEN_merged_up[PTEN_merged_up['prediction'] == 'Neutral'])

act_LOF = len(PTEN_merged_up[(PTEN_merged_up['Polyphen2_HVAR_pred'] == 'D') | (PTEN_merged_up['Polyphen2_HVAR_pred'] == 'P')])
pre_LOF = len(PTEN_merged_up[PTEN_merged_up['prediction'] == 'LOF'])

pre_GOF = len(PTEN_merged_up[PTEN_merged_up['prediction'] == 'GOF'])

categories = ['Neutral', 'LOF', 'GOF']
actual = [act_neutral, act_LOF, 0]
predicted = [pre_neutral, pre_LOF, pre_GOF]
```

```python
bar_width = 0.2  # Width of each bar
bar_positions_set1 = range(len(categories))
bar_positions_set2 = [pos + bar_width for pos in bar_positions_set1]
bar_positions_set3 = [pos + bar_width for pos in bar_positions_set2]

plt.barh(bar_positions_set1, actual, height=bar_width, label='Experimental')
plt.barh(bar_positions_set2, predicted, height=bar_width, label='Predicted')

plt.xlabel('Number of Variants')
plt.legend()

# Set y-axis ticks and labels to categories
plt.yticks([pos + bar_width for pos in bar_positions_set1], categories)

# Remove y-axis tick marks
plt.tick_params(axis='y', which='both', length=0)

plt.savefig('/Users/sayehkohani/Downloads/PTEN_Logofunc.png', bbox_inches='tight')  # Change the filename and format as needed

plt.show()
```
